# Accuracy tuning using `max-tree-size`

As explain in [1,2], the only difference between the MWPM decoder and the Union-Find decoder is whether they maintain detailed structures within each cluster. The `max_tree_size` parameter allows one to control how much detail they want to preserve inside each cluster. When `max_tree_size = 0`, it's equivalent to UF decoder; when `max_tree_size = infinity`, it's equivalent to MWPM decoder. This parameter actually generates a full spectrum of decoders between MWPM and UF decoder. We let the logical error rate of MWPM decoder be $p_M$ and the logical error rate of UF decoder be $p_U$, generally $p_U > p_M$. By tuning `max_tree_size`, we can create a decoder with logical error rate of $p_L$ which has $p_U \geq p_L \geq p_M$.

In Figure 1, we observe that the logical error rate $p_L$ plateaus as `max_tree_size` increases beyond $\left \lceil{(code\_{distance} + 1)/2} \right \rceil $, suggesting that we could obtain the same approximate decoding accuracy of MWPM without setting `max_tree_size` as infinity. The same trend is also observed for planar surface code with perfect stabilizer measurement in Figure 2. 



````python
## For max-half-weight = 128, we plot the logical error rate against max tree size for various code distance 
## circuit level noise: rotated planar code with stim noise model
## Note that $log_2 (max-half-weight * 2)$ represents the number of qubits used

d_vec = range(3,16,2) # code distance
p = 0.0025 # physical error rate
max_tree_size_vec = range(0,12)

for d in d_vec:
    filename = "./max-tree-max-half-weight-data/max_tree_size_accuracy_circuit_level/" + f"{d}_{p}.txt"
    df = pd.read_csv(filename, sep=' ', skiprows=[0], header=None)
    plt.plot(df.iloc[273:288, 1], df.iloc[273:288,7]/df.iloc[288,7], marker=".", label=f"code distance {d}")
    
plt.legend()
plt.xlabel("Max tree size, p=0.0025")
plt.ylabel("Logical Error Rate ($p_{L}$)")
plt.title("Figure 1. Plot of Logical Error Rate $p_{L}$ against max tree size for Circuit Level noise with physical error rate 0.0025 and max-half-weight 128")
````

````python
## For max-half-weight = 128, we plot the logical error rate against max tree size for various code distance 
## standard planar code
## Note that $log_2 (max-half-weight * 2)$ represents the number of qubits used

d_vec = range(3,16,2) # code distance
p = 0.05 # physical error rate
max_tree_size_vec = range(0,12)

for d in d_vec:
    filename = "./max-tree-max-half-weight-data/max_tree_size_accuracy_code_capacity/" + f"{d}_{p}.txt"
    df = pd.read_csv(filename, sep=' ', skiprows=[0], header=None)
    plt.plot(df.iloc[273:288, 1], df.iloc[273:288,7]/df.iloc[288,7], marker=".", label=f"code distance {d}")
    
plt.legend()
plt.xlabel("Max tree size, p=0.05")
plt.ylabel("Logical Error Rate ($p_{L}$)")
plt.title("Figure 2. Plot of Logical Error Rate $p_{L}$ against max tree size for standard planar code with physical error rate 0.05 and max-half-weight 128")
````

# Accuracy tuning with respect to both max_half_weight and max_tree_size

Using `tutorial/max-tree-max-half-weight-data/max_tree_size_accuracy_circuit_level/run.py`, we obtain the logical error rate with respect to both the `max_half_weight` and the `max_tree_size` for circuit level noise, as shown in Figure 2. The bottom left corner corresponds to the weighted union-find algorithm, the bottom right corner the weighted MWPM algorithm, the top left corner unweighted union-find algorithm, and the top right corner the unweighted MWPM. We notice that 
* in general, weighted union-find gives lower logical error rate than unweighted MWPM,
* there are distinct transition lines (vertical) at odd values of `max_tree_size` up until the value of code distance. For instance, for code distance 7 (Fig. 3), we observe distinct lines at `max_tree_size = 3, 5, 7`. Similarly, for code distance 9 (Fig. 4), we observe distinct lines at `max_tree_size = 3, 5, 7, 9`. We attribute these lines to the odd number of nodes in alternating trees in the fast MWPM fusion blossom algorithm.

![title](labeled_heatmap.png)

````python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

## circuit level code type
d_vec = [7, 9] # code distance
p = 0.0025
max_tree_size_vec = range(0,12)
figure, axis = plt.subplots(2, 1, figsize=(7,10))
cmap = sns.cm.rocket_r

for i in range(0,len(d_vec)):
    d = d_vec[i]
    filename = "./max-tree-max-half-weight-data/max_tree_size_accuracy_circuit_level/" + f"{d}_{p}.txt"
    df = pd.read_csv(filename, sep=' ', skiprows=[0], header=None)
    # display(df)
    data = df.pivot(index=df.columns[0], columns=df.columns[1], values=df.columns[7])
    ax = sns.heatmap(data, cbar_kws={'label': f'Logical Error Rate for code distance {d}'}, cmap = cmap, ax=axis[i])
    # plt.legend()


    ax.set_xlabel("Max tree size, p=0.0025")
    ax.set_ylabel("Max half weight, $2^{qubit\_{num}}$")
    ax.set_title(f"Figure {i+3}. " + "Heatmap on Logical Error Rate $p_{L}$ against max tree size and max half weight for code distance " + f"{d}")

plt.show()
````
